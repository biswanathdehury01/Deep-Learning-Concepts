```python
import tensorflow as tf
import numpy as np
import pandas as pd
from pylab import rcParams
import matplotlib.pyplot as plt
import warnings
```

```python
%matplotlib inline
warnings.filterwarnings("ignore")
pd.set_option('display.expand_frame_repr', False)
rcParams['figure.figsize'] = 16, 8
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
# X is input and Y is ouput
```

```python
plt.scatter(df["X"], df["y"])
plt.show()
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1, input_shape=[1])) #  input_shape=[1] is 1 because we have only 1 input feature i.e X
model.compile(loss='mean_squared_error', 
              optimizer=tf.keras.optimizers.Adam(0.01))
```

```python
model.summary()
```

```python
history = model.fit(df["X"], df["y"], epochs=300)
# if you do this in GPU it will be faster
```

```python
plt.plot(history.history['loss'])
plt.show()
```

```python
df["prediction"] = model.predict(df['X'])
```

```python
plt.scatter(df["X"], df["y"])
plt.plot(df["X"], df["prediction"], color='r')
plt.show()
```
