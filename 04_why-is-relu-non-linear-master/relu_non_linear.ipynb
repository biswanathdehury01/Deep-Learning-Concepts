```python
import tensorflow as tf
import numpy as np
import pandas as pd
from pylab import rcParams
import matplotlib.pyplot as plt
import warnings
from mlxtend.plotting import plot_decision_regions
from matplotlib.colors import ListedColormap
```

```python
zero_one_colourmap = ListedColormap(('red', 'green'))
rcParams['figure.figsize'] = 14, 7
```

```python
df = pd.read_excel("Data.xlsx")
```

```python
X = df.loc[:, ["X1", "X2"]].values
y = df.loc[:, "Y"].values
```

```python
plt.scatter(X[:,0],X[:,1], c=y, s=100, cmap = zero_one_colourmap)
plt.show()
```

```python
def create_model(X, y, hidden_neurons, activation, total_epochs):
    # 1.  Create the instance of a Model
  model = tf.keras.models.Sequential()
    # 2. Then create the hidden dense layer
  model.add(tf.keras.layers.Dense(hidden_neurons, # Total no. of neurons in the dense layer, plugged from parameter passed
                                  input_dim=2, # Input Dimentions is 2, as we have only 2 input features in data set
                                  activation=activation)) # Plugged from parameter passed
    # 3. Then we will add final output layer, with a custom activation function
  model.add(tf.keras.layers.Dense(1, activation=activation))
    # 4. We are passing loss funciton, optimizer and metrics by which we will track our model/classifier
  model.compile(loss='binary_crossentropy',
                optimizer='adam',
                metrics=['binary_accuracy'])
    # 5. Then we will fit the training data
  history = model.fit(X, y, epochs=total_epochs)
  return model
```

# Linear Model with 8 Neurons

```python
# here will learn why Linear activation functions are not good option when we have non linearly seperably data

linear_model = create_model(X, y, 8, 'linear', 200)

# Below we can accuracy is not good, as after 200 epoch we have only 40-45% accuracy
```

```python
plot_decision_regions(X, y.ravel(), clf = linear_model, legend=2)
plt.show()

# we can see below very poorly classified
```

# Linear Model with 64 Neurons

```python
linear_model_64_neurons = create_model(X, y, 64, 'linear', 200)

# we can see below accuracy is slighly increased by increasing the number of hidden Neurons.
```

```python
plot_decision_regions(X, y.ravel(), clf = linear_model_64_neurons, legend=2)
plt.show()

# Performance increases as hidden Neurons increased
```

# Relu Activation with 8 neurons

ReLU is a non linear Activation function

```python
relu_8_neuron_model = create_model(X, y, 8, 'relu', 200)

# we can see with few hidden Neurons and ReLU af, we can see the model is not able to find the pattern in the data
```

```python
plot_decision_regions(X, y.ravel(), clf = relu_8_neuron_model, legend=2)
plt.show()

# boundaries are not accuracte, but better than linear af.
```

# Relu Activation with 64 neurons

```python
relu_64_neuron_model = create_model(X, y, 64, 'relu', 200)

# Now lets increase the nuerons to 64, we can see accuracy reacher maximum.
```

```python
plot_decision_regions(X, y.ravel(), clf = relu_64_neuron_model, legend=2)
plt.show()
```
